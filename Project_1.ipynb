```python
import pandas as pd
us= pd.read_csv('USvideos.csv')
uk= pd.read_csv('GBvideos.csv')
india= pd.read_csv('INvideos.csv')
```

```python
us["country"]="US"
uk["country"]="UK"
india["country"]="India"
df = pd.concat([us, india, uk], ignore_index=True)
print(df.shape)
```

```python
df['trending_date'].head(10)
```

```python
df['trending_date'] = pd.to_datetime(df['trending_date'], format='%y.%d.%m')
```

```python
df['trending_date'].min(), df['trending_date'].max()
```

```python
df['publish_time'].min(), df['publish_time'].max()
```

```python
df = df.drop_duplicates()
df.shape
```

```python
(df['views'] == 0).sum()
```

```python
df['engagement_rate'] = (df['likes'] + df['comment_count']) / df['views']
df['like_ratio'] = df['likes'] / df['views']
df['comment_ratio'] = df['comment_count'] / df['views']
```

```python
df['publish_time'].dtype
```

```python
df['publish_time'] = pd.to_datetime(df['publish_time'])
df['publish_time'].dtype
```

```python
df['publish_hour'] = df['publish_time'].dt.hour
df['publish_day'] = df['publish_time'].dt.day_name()
df['publish_month'] = df['publish_time'].dt.month
```

```python
trend_duration = df.groupby(['video_id','country'])['trending_date'].nunique()
trend_duration.describe()
```
